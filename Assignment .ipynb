# BINOMIAL TREE 

```python
import os

#this package imports fonts 
```

data input: 

```python
S0 = 160.52

#spot stock price, from historical data (adjusted closed price 29.10.2021)

K = 170.0

#strike price. 

T = 0.25

#3 month maturity.

r = 0.0155

#10 years risk free rate on 29.10.2021.

vol = 0.15

#annualized volatility.

N = 6

#number of periods or number of time steps. 

payoff = "put"          

#payoff 
```

```python
dT = float(T) / N

# Delta T

u = np.exp(vol * np.sqrt(dT))

# up factor

d = 1.0 / u 

# down factor 
```

Create a tree: 

```python
S = np.zeros((N + 1, N + 1))
S[0, 0] = S0
z = 1                         

#z is a pointer variable, and it creates a range

for t in range(1, N + 1):
    for i in range(z):
        S[i, t] = S[i, t-1] * u
        S[i+1, t] = S[i, t-1] * d
    z += 1
    
#This code loops a dinamic 4 by 4 matrix 

S
```

```python
a = np.exp(r * dT)    

#risk free compound return

p = (a - d)/ (u - d) 

#risk neutral up probability

p.round(4)
```

```python
q = 1.0 - p 

#risk neutral down probability. 1.0 is used to make sure the result is a float number.

q.round(4)
```

Option value at each final node: 

```python
S_T = S[:,-1]
V = np.zeros((N + 1, N + 1))
if payoff =="call":
    V[:,-1] = np.maximum(S_T-K, 0.0)
elif payoff =="put":
    V[:,-1] = np.maximum(K-S_T, 0.0)
V

#S_T is the spot price of the underlying asset at the T-th period.
#this code creates a matrix with the underlying asset price
```

The value V is the payoff value at maturity. 

European Option: 
    

```python
for j in range(N-1, -1, -1):
    for i in range(j+1):
        V[i,j] = np.exp(-r*dT) * (p * V[i,j + 1] + q * V[i + 1,j + 1])
V


#this code generates the matrix value for European Option, it has two loops, one for the column and one for the row
```

```python
print('European ' + payoff, str( V[0,0]))
```

# Monte Carlo Simulation 


the Monte Carlo Simulation has only one loop of code, over the time interval, the other dimension is a vectorised code for all the path. 

```python
def mcs_simulation_np(p):
    M = p
    I = p
    
#p represents the time steps 

    dt = T / M 
    S = np.zeros((M + 1, I))
    S[0] = S0 
    rn = np.random.standard_normal(S.shape) 
    for t in range(1, M + 1): 
        S[t] = S[t-1] * np.exp((r - vol ** 2 / 2) * dt + vol * np.sqrt(dt) * rn[t]) 
        
#S[t] = today's price 
#S[t-1] = yesterday's price 


    return S
```

data inputs:

```python
T = 0.25

#time to maturity 

r = 0.0155

#10 years risk free rate on 29.10.2021.

vol = 0.15 

#Volatility.

S0 = 160.52

#spot stock price, from historical data (adjusted closed price 29.10.2021)

K = 170.00

#Strike price.
```

Delta: 
    

```python
def delta(S, K, T, r, q, vol, payoff):
    
#q: continuous dividend yield = 0.0268 (data from yahoo finance)
    
    d1 = (np.log(S / K) + (r - q + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    if payoff == "call":
        delta = np.exp(-q * T) * si.norm.cdf(d1, 0.0, 1.0)
    elif payoff == "put":
        delta =  - np.exp(-q * T) * si.norm.cdf(-d1, 0.0, 1.0)
    
    return delta
```

Gamma: 

```python
def gamma(S, K, T, r, q, vol, payoff):
    
    d1 = (np.log(S / K) + (r - q + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    gamma = np.exp(- q * T) * si.norm.pdf(d1, 0.0, 1.0) / (vol * S * np.sqrt(T))
    
    return gamma
```

Theta:

```python
def theta(S, K, T, r, q, vol, payoff):
    
    d1 = (np.log(S / K) + (r - q + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    d2 = (np.log(S / K) + (r - q - 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    if payoff == "call":
        theta = vol * S * np.exp(-q * T) * si.norm.pdf(d1, 0.0, 1.0) / (2 * np.sqrt(T)) - q * S * np.exp(-q * T) * si.norm.cdf(d1, 0.0, 1.0) + r * K * np.exp(-r * T) * si.norm.cdf(d2, 0.0, 1.0)
    elif payoff == "put":
        theta = vol * S * np.exp(-q * T) * si.norm.pdf(-d1, 0.0, 1.0) / (2 * np.sqrt(T)) - q * S * np.exp(-q * T) * si.norm.cdf(-d1, 0.0, 1.0) + r * K * np.exp(-r * T) * si.norm.cdf(-d2, 0.0, 1.0)
    
    return theta
```

Rho: 

```python
def rho(S, K, T, r, q, vol, payoff):
    
    d1 = (np.log(S / K) + (r - q + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    d2 = (np.log(S / K) + (r - q - 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    if payoff == "call":
        rho =  K * T * np.exp(-r * T) * si.norm.cdf(d2, 0.0, 1.0)
    elif payoff == "put":
        rho = - K * T * np.exp(-r * T) * si.norm.cdf(-d2, 0.0, 1.0)
    
    return rho
```

Vega:

```python
def vega(S, K, T, r, q, vol, payoff):
    
    d1 = (np.log(S / K) + (r - q + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    vega = S * np.sqrt(T) * np.exp(-q * T) * si.norm.pdf(d1, 0.0, 1.0)
    
    return vega
```

```python
vega(160.52, 170, 0.25, 0.0155, 0.0268, 0.15, 'put').round(4)
```
